```python
def plot_correlation_map( df ):
    corr = df.corr()
    s , ax = mlp.subplots( figsize =( 12 , 10 ) )

    cmap = sb.diverging_palette( 220 , 10 , as_cmap = True )

    s = sb.heatmap(corr, cmap = cmap, square=True, 

    cbar_kws={ 'shrink' : .9 }, 

    ax=ax, 

    annot = True, 

    annot_kws = { 'fontsize' : 12 })
```

```python
plot_correlation_map(titianic2)
#plottinng the correlation map of the dataset
```

```python
Old_title=[]
New_title=[]


for name in titianic2['Name']:
    for m,o in zip(Mo_dict.keys(), Mo_dict.values()):
        m= m+ "."
        if m in name:
            Old_title.append(m)
            New_title.append(o)
```
