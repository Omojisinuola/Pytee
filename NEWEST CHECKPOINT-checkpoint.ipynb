```python
FS= House[['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'waterfront',
              'view', 'condition', 'grade', 'sqft_above', 'sqft_basement', 'yr_built',
              'yr_renovated', 'lat', 'long', 'sqft_living15', 'sqft_lot15', 'price']]
```

```python
Sc= StandardScaler()
trans= Sc.fit_transform(FS)
```

# APPLYING LINEAR REGRESSION TO FEATURE YEAR BUILT

```python
x_yr_built= trans_df.iloc[:, 11]
y_yr_built= trans_df.iloc[:, -1]
```

```python
x_train_yr_built, x_test_yr_built, y_train_yr_built, y_test_yr_built= train_test_split(x_yr_built, y_yr_built, test_size = .2)
```

```python
x_train_yr_built= np.array(x_train_yr_built).reshape(-1, 1)
x_test_yr_built= np.array(x_test_yr_built).reshape(-1, 1)
```

```python
MJ= LinearRegression()
MJ.fit(x_train_yr_built, y_train_yr_built)

target= MJ.predict(x_test_yr_built)
```

```python
print('MSE:', mean_squared_error(y_test, target))

print('R_SQUARED:', metrics.r2_score(y_test, target))
```

# BUILDING A MODEL AND VISUALISING USING SIMPLE LINEAR REGRESSION

```python
import matplotlib.pyplot as mlp #IMPORTING MY MATPLOT LIBRARY.
```

```python
Lin_reg= LinearRegression()
```

```python
Lin_reg.fit(x_train_yr_built, y_train_yr_built)
```

```python
print('Intercept constant for our linear model:', Lin_reg.intercept_)
print('coefficient of x for our linear model:', Lin_reg.coef_)
```

```python
y_pred= Lin_reg.predict(x_train_yr_built)
```

```python
mlp.scatter(x_train_yr_built, y_train_yr_built, color= 'red')
mlp.plot(x_train_yr_built, y_pred, color= 'black')
mlp.title('Truth or Bluff (Linear Regression)')
mlp.xlabel('yr_built')
mlp.ylabel('price')
mlp.show()
```

# TRAINING THE POLYNOMIAL REGRESSION MODEL ON THE WHOLE DATASET

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg= PolynomialFeatures(degree= 4)
```

```python
x_poly= poly_reg.fit_transform(x_train_yr_built)
lin_reg_2= LinearRegression()
lin_reg_2.fit(x_poly, y_train_yr_built)
```

# VISUALISING THE POLYNOMIAL REGRESSION RESULTS

```python
y_poly_pred= lin_reg_2.predict(x_poly)
y_poly_pred
```

```python
mlp.scatter(x_train_yr_built, y_train_yr_built, color= 'red')
mlp.plot(x_train_yr_built, y_poly_pred, color= 'black')
mlp.title('Truth or Bluff (Polynomial Regression)')
mlp.xlabel('yr_built')
mlp.ylabel('price')
mlp.show()
```
